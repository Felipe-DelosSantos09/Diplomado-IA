```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Datos de entrenamiento para la compuerta XOR
input_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
target_output = np.array([0, 1, 1, 0])

# Crear el modelo del perceptrón multicapa
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))  # Capa oculta con 4 neuronas y función de activación ReLU
model.add(Dense(1, activation='sigmoid'))           # Capa de salida con 1 neurona y función de activación sigmoide

# Compilar el modelo
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy']) # loss='binary_crossentropy', optimizer='adam', metrics=['acc']

# Entrenar el modelo
epochs = 2500
history = model.fit(input_data, target_output, epochs=epochs, verbose=2)


```

```python
# Evaluar el modelo entrenado
loss, accuracy = model.evaluate(input_data, target_output)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

```python
# Realizar predicciones
test_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predictions = model.predict(test_inputs)
print("Predictions:")
for i in range(len(test_inputs)):
    print(f"Input: {test_inputs[i]}, Prediction: {predictions[i][0].round()}")
```
