```python
food_clean = food.drop("Eutrophying emissions per 100g protein (gPO₄eq per 100 grams protein)",axis=1) 
```

```python
food_clean["Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)"] = food_clean["Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)"].fillna(food_clean["Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)"].mean())
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

impute_numeric = KNNImputer()
impute_numeric = SimpleImputer(strategy="mean")
numeric_colums = food_clean.select_dtypes(include=["number"]).columns
numeric_colums 
```

```python
food_clean[numeric_colums] = impute_numeric.fit_transform(food_clean[numeric_colums])
```

```python
impute_categorico = SimpleImputer(strategy="most_frequent")
categorical_colums = food_clean.select_dtypes(include=["object"]).columns
food_clean[categorical_colums] = impute_categorico.fit_transform(food_clean[categorical_colums])
```

```python
food_clean.isnull().sum()
```

```python
for name in food_clean.columns.to_list():
    print(f"{name} : {food_clean[name].unique()}")
    print("\n")
```
