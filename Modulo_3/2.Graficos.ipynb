```python
experiencia = [1, 3, 5, 7, 9, 13]
salario_data_scientists = [9000, 1200, 15000, 20000, 25000, 40000]
salario_ing_software  = [8000, 1100, 13000, 18000, 22000, 35000]
```

```python
plt.plot(experiencia,salario_data_scientists)
plt.plot(experiencia,salario_ing_software)
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.show()
```

```python
plt.plot(experiencia,salario_data_scientists)
plt.plot(experiencia,salario_ing_software)
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.legend(["Data Scientist", "Ing Software"])
plt.show()
```

```python
plt.plot(experiencia,salario_data_scientists, color = "r", label = "Data Scientist")
plt.plot(experiencia,salario_ing_software, color = "g", label = "Ing Software")
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.legend()
plt.show()
```

```python
plt.plot(experiencia,salario_data_scientists, color = "r", label = "Data Scientist", linestyle = "--", marker = "o")
plt.plot(experiencia,salario_ing_software, color = "g", label = "Ing Software", linestyle = ":", marker = "*")
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.legend()
plt.grid(color = "r", linestyle = "-", linewidth = 2)
plt.show()
```

```python
x = ["A", "B", "C", "D"]

y = [3, 8, 10, 1]

plt.bar(x,y)
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.show()
```

```python
x = ["A", "B", "C", "D"]

y = [3, 8, 10, 1]

tendencia = [3, 8, 10, 1]

plt.bar(x,y)
plt.plot(x,tendencia, color = "r")
plt.title("Salarios")
plt.xlabel("Experiencia")
plt.ylabel("Ganancias")
plt.show()
```

```python
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x,y, color = "g", alpha = 0.5)
plt.title("Grafico de dispersión")
plt.show()
```

```python
datos = np.random.randn(10000)

plt.hist(datos,bins=10)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(datos,vert=True)
plt.show()
```

```python
plt.xlabel("Valor")
```

```python
import pandas as pd

data_range = pd.date_range(start="2023-01-01", end = "2023-12-31", freq="D")
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({"Date":data_range, "Precios":precios})
data.head()
```

```python
plt.plot(data["Date"], data["Precios"], label="Precio")
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

categorias = ["a", "b", "c", "d"]
valores = [5,8,2,4]

x1= np.random.rand(100)
y = np.random.rand(100)

n = np.random.randn(100000)

```

```python
fig, axs = plt.subplots(2,2, figsize = (15,10))

axs[0,0].plot(x,y1, label = "Seno")
axs[0,0].plot(x,y2, label = "Coseno")
axs[0,0].set_title("Grafico de líneas")
axs[0,0].legend(loc="upper right") #center
axs[0,0].grid(True)

axs[0,1].bar(categorias,valores)
axs[0,1].set_title("Grafico de barras")
axs[0,1].grid(True)

axs[1,0].hist(n,bins=100)
axs[1,0].set_title("Histograma")
axs[1,0].grid(True)

axs[1,1].scatter(x1,y, color = "g", alpha = 0.5)
axs[1,1].set_title("Grafico de dispersión")
axs[1,1].grid(True)

fig.suptitle("Ejemplos de diferentes tipos de graficos",fontsize = 25)
plt.show()
```
