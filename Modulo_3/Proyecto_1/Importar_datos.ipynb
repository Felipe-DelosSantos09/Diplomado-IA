```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

impute_numeric = KNNImputer()
impute_numeric = SimpleImputer(strategy="mean")
numeric_colums = df_clean.select_dtypes(include=["number"]).columns
#df_clean[numeric_colums] = impute_numeric.fit_transform()
numeric_colums 
```

## Extraer numeros de una columna

```python
import pandas as pd
import re
import numpy as np
data = {"mezclados": ["4526 ewd", "37437 fkh", "9806 yue", "4763 ire"]}
df = pd.DataFrame(data)
```

```python
data_num = "4526 ewd"

numero = re.search(r"\d+", data_num)
int(numero.group())
```

```python
def extrac_numbers(s):
    numero = re.search(r"\d+", data_num)
    return int(numero.group()) if numero else np.nan
```

```python
df["mezclados"] = df["mezclados"].apply(extrac_numbers)
```

# Detección de valores atipicos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## NOTAS: Simulación de datos atipicos

```python
Q1 = df_clean["Age"].quantile(0.25)
Q3 = df_clean["Age"].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1-5 * IQR

limite_inferior

limite_superior = Q3 + 1.5 * IQR

limite_superior
```

```python
df_test  = df_clean.copy()

indices = np.random.choice(df_test.index, 2, replace=False)

df_test.loc[indices[0], "Age"] += 3 * IQR
df_test.loc[indices[1], "Age"] += 3 * IQR
```

```python
df_test.describe()
```

```python
df_test.boxplot(column="Age")
```

```python
#Calculamos el rango de datos que no son atipicos
Q1 = df_test["Age"].quantile(0.25)
Q3 = df_test["Age"].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR

print(limite_inferior)

limite_superior = Q3 + 1.5 * IQR

print(limite_superior)
```
