# Ejercicio cargar y mostrar una imagen

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# Cargar la imagen
img_path = r'20230423_111633.jpg'  # Cambia esta ruta a la ubicacin de tu imagen
img = image.load_img(img_path)

# Mostrar la imagen
plt.imshow(img)
plt.axis('off')  # Ocultar los ejes
plt.show()
```

## Redimensionar una imagen

```python
# Redimensionar la imagen
img_resized = img.resize((64, 64))

# Mostrar la imagen redimensionada
plt.imshow(img_resized)
plt.axis('off')
plt.show()
```

## Convertir una imagten a un array y normalizarla


```python
import numpy as np

# Convertir la imagen a un array de Numpy
img_array = image.img_to_array(img_resized)

# Normalizar los valores de los pxeles a [0, 1]
img_array = img_array / 255.0

print(f"Array de la imagen normalizado: {img_array}")
```

## Aplicar rotacin a una imagen

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Crear un generador de imgenes
datagen = ImageDataGenerator(rotation_range=90)

# Convertir la imagen a un formato compatible con el generador
img_array = np.expand_dims(img_array, axis=0)

# Generar una imagen rotada
img_rotated = next(datagen.flow(img_array, batch_size=1))[0]

# Mostrar la imagen rotada
plt.imshow(img_rotated)
plt.axis('off')
plt.show()
```

## Aplicar un recorte  aleatorio a la imagen

```python
# Crear un generador de imgenes con recorte
datagen = ImageDataGenerator(width_shift_range=0.2, height_shift_range=0.2)

# Generar una imagen recortada
img_cropped = next(datagen.flow(img_array, batch_size=1))[0]

# Mostrar la imagen recortada
plt.imshow(img_cropped)
plt.axis('off')
plt.show()
```

## Aplicar zoom aleatorio a la imagen recortada

```python
# Crear un generador de imgenes con zoom
datagen = ImageDataGenerator(zoom_range=0.5)

# Generar una imagen con zoom
img_zoomed = next(datagen.flow(img_array, batch_size=1))[0]

# Mostrar la imagen con zoom
plt.imshow(img_zoomed)
plt.axis('off')
plt.show()
```

## Aplicar un desplazamiento horizontal y vertical a la imagen

```python
# Crear un generador de imgenes con desplazamiento
datagen = ImageDataGenerator(width_shift_range=0.2, height_shift_range=0.2)

# Generar una imagen desplazada
img_shifted = next(datagen.flow(img_array, batch_size=1))[0]

# Mostrar la imagen desplazada
plt.imshow(img_shifted)
plt.axis('off')
plt.show()
```

Ampliacin de Datos (Data Augmentation)

La ampliacin de datos es una tcnica utilizada en el aprendizaje automtico y el reconocimiento de imgenes para aumentar la cantidad y la diversidad de los datos de entrenamiento mediante la aplicacin de diversas transformaciones a las imgenes originales. Esto ayuda a mejorar la capacidad del modelo para generalizar a nuevos datos y reduce el sobreajuste (overfitting).

Lo que se vio anteriormente es un ejemplo de esto, sin embargo en una implementacion de con Keras se veria de la siguiente manera

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen
img_path = r'D:\DIPLOMADOS\IA-Diplomado\Modulo 4\woman-8828359_640.jpg' # Cambia esta ruta a la ubicacin de tu imagen
img = image.load_img(img_path)

# Redimensionar la imagen para asegurar un tamao uniforme
img_resized = img.resize((64, 64))

# Convertir la imagen a un array de Numpy
img_array = image.img_to_array(img_resized)
img_array = img_array / 255.0  # Normalizar los valores de los pxeles a [0, 1]

# Expandir las dimensiones de la imagen para que tenga la forma (1, height, width, channels)
img_array = np.expand_dims(img_array, axis=0)

# Crear un generador de imgenes con varias transformaciones
datagen = ImageDataGenerator(
    rotation_range=45,             # Rotacin aleatoria en el rango de 0 a 45 grados
    width_shift_range=0.2,         # Desplazamiento horizontal aleatorio
    height_shift_range=0.2,        # Desplazamiento vertical aleatorio
    zoom_range=0.5,                # Zoom aleatorio
    horizontal_flip=True,          # Volteo horizontal aleatorio
    vertical_flip=True             # Volteo vertical aleatorio
)

# Generar una imagen transformada
img_transformed = next(datagen.flow(img_array, batch_size=1))[0]

# Mostrar la imagen transformada
plt.imshow(img_transformed)
plt.axis('off')
plt.show()
```

## Otras Herramientas para Ampliacin de Datos

Adems de Keras, existen otras herramientas poderosas para la ampliacin de datos, como:

Albumentations
Albumentations es una biblioteca rpida y flexible para la ampliacin de datos de imgenes. Proporciona una amplia variedad de transformaciones y es conocida por su eficiencia. Aqu tienes un ejemplo de cmo usar Albumentations:

```python
import cv2
import albumentations as A
import matplotlib.pyplot as plt

# Cargar la imagen
img_path = r'20230423_111633.jpg'  # Cambia esta ruta a la ubicacin de tu imagen
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python


# Definir las transformaciones
transform = A.Compose([
    A.RandomRotate90(),         # Rotacin aleatoria de 90 grados
    A.Flip(),                   # Volteo horizontal y vertical
    A.Transpose(),              # Transponer la imagen
    A.OneOf([
        A.MotionBlur(p=0.2),    # Desenfoque de movimiento
        A.MedianBlur(blur_limit=3, p=0.1),  # Desenfoque mediano
        A.Blur(blur_limit=3, p=0.1),        # Desenfoque
    ], p=0.2),
    A.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.2, rotate_limit=45, p=0.2),
    A.RandomBrightnessContrast(p=0.2),  # Cambio aleatorio de brillo y contraste
    A.HueSaturationValue(p=0.3),  # Cambio aleatorio de hue, saturacin y valor
])

# Aplicar las transformaciones
transformed = transform(image=img)
img_transformed = transformed['image']

# Mostrar la imagen transformada
plt.imshow(img_transformed)
plt.axis('off')
plt.show()
```
